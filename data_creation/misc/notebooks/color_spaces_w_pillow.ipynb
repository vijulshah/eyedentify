```python
from PIL import Image
import os
from torchvision import transforms
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_images(images, titles, figsize=(12, 8)):
    plt.figure(figsize=figsize)
    for i, (image, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        if isinstance(image, Image.Image):  # Check if the image is a PIL Image
            image = np.array(image).transpose(1, 2, 0)
        if len(image.shape) == 2:  # Check if the image is grayscale
            plt.imshow(image, cmap='gray')
        else:
            plt.imshow(image)
        plt.title(title)
        plt.axis('off')
    plt.show()
```

```python
# Define the path to your image
data_path = "/home/vshah/Documents/Uni/Thesis_Cluster/pupil-size-estimation-with-super-resolution/data/EyeDentify/W_SR/GFPGAN_x2/left_eye/25/1/frame_01.png.png"

save_path = "/home/vshah/Documents/Uni/Thesis_Cluster/pupil-size-estimation-with-super-resolution/preprocessing"

preprocess_function = transforms.Compose([transforms.ToTensor()])
```

```python
# Open the image
image = Image.open(os.path.join(data_path))

rgb_image = image.convert("RGB")
rgb_image_tensor = preprocess_function(rgb_image)
print(rgb_image_tensor.shape)

rgb_image_tensor = rgb_image_tensor.numpy().transpose(1, 2, 0)
print(rgb_image_tensor.shape)
```

```python
# Convert the image to YCbCr mode
YCbCr_image = image.convert("YCbCr")
YCbCr_image_tensor = preprocess_function(YCbCr_image)
print(YCbCr_image_tensor.shape)

YCbCr_image_tensor = YCbCr_image_tensor.numpy().transpose(1, 2, 0)
print(YCbCr_image_tensor.shape)
```

```python
# Convert the image to HSV mode
hsv_image = image.convert("HSV")
hsv_image_tensor = preprocess_function(hsv_image)
print(hsv_image_tensor.shape)

hsv_image_tensor = hsv_image_tensor.numpy().transpose(1, 2, 0)
print(hsv_image_tensor.shape)
```

```python
# Convert the image to LAB mode
lab_image = image.convert("LAB")
lab_image_tensor = preprocess_function(lab_image)
print(lab_image_tensor.shape)

lab_image_tensor = lab_image_tensor.numpy().transpose(1, 2, 0)
print(lab_image_tensor.shape)
```

```python
# Convert the image to grayscale mode
gray_image = image.convert("L")
gray_image_tensor = preprocess_function(gray_image)
print(gray_image_tensor.shape)

gray_image_tensor = gray_image_tensor.squeeze(0)
gray_image_tensor = gray_image_tensor.numpy()
print(gray_image_tensor.shape)
```

```python
show_images([rgb_image_tensor, gray_image_tensor, hsv_image_tensor, lab_image_tensor, YCbCr_image_tensor], ["RGB", "Grayscale", "HSV", "LAB", "YCbCr"])
```
